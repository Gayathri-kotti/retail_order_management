```python
import numpy as np
import pandas as pd
```

```python
# read data
df=pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
df.head(20)
```

```python
# handle null values 
#  to handle those first lets see how these null values are named as some are named as not available/unknown,etc
df['Ship Mode'].unique() # here nan is fine , we need to convert not available/unknown as nan
#  output is this when df=pd.read_csv('orders.csv')
```

```python
df['Ship Mode'].unique() 
#  output is this when df.pd.read_csv('orders.csv',na_values=['Not Available','unknown']) -> na_values converts this(not available/unknown) into nan
```

```python
# rename the column names ..make them lower case and replace space with underscore that helps in retreiving the data without raising a keyerror
df.rename(columns={'Order Id':'order_id','City':'city'}) # it can be bothered because we should write this for all columns
```

```python
# instead we can do this to rename all column names
#  step -1: retreive all col names -> df.columns
# step-2: convert it into a string object and use lower() string method
df.columns=df.columns.str.lower()
print(df.columns)
# now we should handle underscore , we can use replace()
df.columns=df.columns.str.replace(' ','_')
print(df.columns)
```

```python
df.head()
```

```python
# derive new column as discount, sale profit and profit
df['discount']=df['list_price']*df['discount_percent']*0.1
df['sale_price']=df['list_price']-df['discount']
df['profit']=df['sale_price']-df['cost_price']
df.head()
```

```python
# convert order date from object data type to datetime
# df.dtypes
df['order_date']=pd.to_datetime(df['order_date'],format="%Y-%m-%d")
df.dtypes
```

```python
# to drop any columns in df 
# df.drop(columns=['list_price','discount_percent','cost_price'])
#  the above code drops the columns but does not aplly the changes to original df so we inplace =true
df.drop(columns=['list_price','discount_percent','cost_price'],inplace=True)
```
